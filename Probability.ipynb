3.finding the dataset of june and july with floods

```python
df["JUN_GT_500"]=(df["JUN"]>500).astype("int")
df["JUL_GT_500"]=(df["JUL"]>500).astype("int")
df["FLOODS"]=df["FLOODS"].map({"YES":1,"NO":0})
df["FLOODS"].head()
df2=df.loc[:,["YEAR","JUN_GT_500","JUL_GT_500","FLOODS"]]
df2["COUNT"]=1
df2.head()
```

```python
#shape
df2.shape
```

```python
#using crosstab
pd.crosstab(df2['FLOODS'],df2["JUN_GT_500"])
```

```python
#finding the probability of flood,june,and intersection
pro_flood=(6+54)/(6+54+19+39)
print(pro_flood)
pro_june=(39+54)/(6+54+19+39)
print(pro_june)
intersection_june_flood=(54)/(6+39+19+54)
print(intersection_june_flood)
```

4.calculating the (p(A|B)) of june

```python
pro_jun=intersection_june_flood/pro_june
print("Probability  of  rain of more  than  500  mm  in  June  given  it  flooded  that  year (P(A|B)):")
print("p(flood|june):",pro_jun)
```

5.calculating the(p(B|A))of june

```python
pro_jun_flo=(pro_jun*pro_june)/pro_flood
print("Probability  of  rain of more  than  500  mm  in  June  given  it  flooded  that  year (P(B|A)):")
print("p(june|flood):",pro_jun_flo)
```
